```python
import pandas as pd

data = {
    "skills": [
        "python machine learning data analysis",
        "java spring boot backend development",
        "html css javascript react frontend",
        "python deep learning tensorflow keras",
        "sql database excel powerbi"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df['skills'] = df['skills'].str.lower()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english')
X = tfidf.fit_transform(df['skills'])
```

```python
def match_resume(resume):
    vec = tfidf.transform([resume])
    sim = cosine_similarity(vec, X)
    return sim.max()

def skill_gap(user, job):
    return list(set(job.split()) - set(user.split()))
```

```python
print("AI Job Analyzer")
print("----------------")

resume = input("Enter your skills: ")
job = input("Enter job description: ")

print("\nMatch Score:", match_resume(resume))
print("Missing Skills:", skill_gap(resume, job))
```
